```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
#data, target = load_boston(return_X_y = True)

#1) load boston dataset
boston = load_boston()
data = boston.data
target = boston.target
df = pd.DataFrame(data)
df.columns = boston.feature_names
df["PRICE"] = target
df.head()
```

```python
#2) Normalize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#scaler.fit(data)  #Identifica a função que normaliza cada coluna
data_normalized = scaler.fit_transform(data)    #Identifica a função que normaliza e ja transforma os dados
df_norm = pd.DataFrame(data_normalized)
df_norm.columns = boston.feature_names
df_norm["PRICE"] = target

df_norm.head()
```

```python
#3
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
parameters = {'kernel': ('rbf', 'sigmoid'), 'C': [5, 10]}
svr = SVR(gamma = 'auto')
grid = GridSearchCV(
    svr,
    parameters,
    cv = 5,
    scoring = 'neg_mean_squared_error',
#     return_train_score = True,
#     refit = True
)
#grid.fit(data, target)
grid.fit(data_normalized, target)
grid.best_params_
```

```python
from sklearn.neural_network import MLPRegressor
parameters = {'activation': ['identity', 'logistic', 'tanh', 'relu'],
              'solver': ['lbfgs', 'sgd', 'adam'],
              'hidden_layer_sizes': (20, 10)}
mlp = MLPRegressor()
grid_reg = GridSearchCV(
    mlp,
    parameters,
    cv = 5,
    scoring = 'neg_mean_squared_error'
)

grid_reg.fit(data_normalized, target)
grid_reg.best_params_
```
